# Testing MATLAB Jupyter Notebook Functionality

It is pretty awesome that the notebook environment is compatible now with MATLAB. This will allow us to mix theoretical and programmatic notes in one fell swoop. For instance, say we want to expose students to plotting in matlab. we could create an instructional document that explains:

$$y = \sin(t)$$ 

in some meaningful way, and then shows how to plot this equation in code.

```matlab
t = linspace(0,10,1000);
y = sin(t);
plot(t,y,'k')
xlabel('Time (s)')
ylabel('output')
```

## Application example: Controls

Additionally, using the "cell magics" feature of the Jupyter notebook, it is fairly easy to do some pretty wild things to explain concepts to students using a mixture of code, graphics, and text/equations. For example, to insert an image into the notebook, we could do something like:



```matlab
[im, map, alpha] = imread('controller_diagram.png');
f=imshow(im);
set(f, 'AlphaData', alpha);
```


Then, we could use this diagram to explain a bit of code, like saying that if the system is first order, of the form:

$$G(s) = \frac{1}{s+1}$$

and the controller has the form 

$$C(s) = kp+\frac{ki}{s}$$

Then the closed-loop transfer function of the system is 

$$G_{cl}(s) = \frac{G(s)C(s)}{1+G(s)C(s)}$$

And in matlab, we can look at the system's step response for $kp=1$ and $ki=2$ fairly easily.

```matlab
s = tf('s');
G = 1/(s+1);
C = 1 + 2/s;

Gcl = G*C/(1+G*C)

step(Gcl)
```
