# Exploring the Design Space re: Motion Envelope

If you have seen the finding_envelope.ipynb notebook, you will notice that we went through each of the 6 degrees of freedom for the motion base and determined minimum and maximum values. I have taken the liberty of adding this functionality to the motionbase.py function in the MotionBase class. This means that when you initialize a motion base with particular dimensions, the envelope is calculated right away, and variables like base.xmin, base.yawmin, etc. are calculated and ready for you to call. This could be very helpful when actually controlling the base, to ensure that values you ask the base to go to are within acceptable limits. The function in the MotionBase class also resets the "resting" height of the base to the true middle height.

We will be using this new functionality to look at how design choices in the motion base size will affect its achievable motion envelope.

```python
%matplotlib inline
from numpy import *
from motionbase import MotionBase


# we will come up with a vector of motion base sizes to test. 
baseradius = 0.5 #meters
#set conrod and motor arm length
motorarm = 0.3
conrod = 1.0
#look at a range of platform radii
platformradius = baseradius*linspace(0.5,1,50)

#maybe look just at how Z range is affected to start
zmax = array([])
for ind in range(0,len(platformradius)):
    base = MotionBase(1.0,baseradius,platformradius[ind],motorarm,conrod)
    zmax = append(zmax,base.zmax)

plot(platformradius/baseradius,zmax/baseradius)
xlabel('normalized platform size')
ylabel('normalized maximum height')
```
